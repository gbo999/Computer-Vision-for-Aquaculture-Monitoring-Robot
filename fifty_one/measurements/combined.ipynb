```python
# First, extract the camera/source from the image names
df_combined['source'] = df_combined['image_name'].str.extract(r'(GX\d+)')

# Create separate thresholds for each source
thresholds = {}
for source in df_combined['source'].unique():
    source_data = df_combined[df_combined['source'] == source]
    
    big_mean = source_data['Big_pixels_total_length'].mean()
    small_mean = source_data['Small_pixels_total_length'].mean()
    
    # Only calculate threshold if we have both measurements
    if not (pd.isna(big_mean) or pd.isna(small_mean)):
        thresholds[source] = (big_mean + small_mean) / 2
    
print("Thresholds by source:", thresholds)

# Function to classify based on source and pixel length
def classify_by_source(row):
    source = row.get('source')  # Get the camera source
    length = row.get('Length')  # Replace with your actual column name
    
    if pd.isna(length) or source not in thresholds:
        return None
    elif length > thresholds[source]:
        return 'big'
    else:
        return 'small'

# Now for your df_shai, assuming it has similar structure:
# First extract the source from image names
if 'Label' in df_shai.columns:
    df_shai['source'] = df_shai['Label'].str.extract(r'(GX\d+)')
    
    # Apply the classification
    df_shai['size_class'] = df_shai.apply(classify_by_source, axis=1)
    
    # Check the results
    print(df_shai.groupby(['source', 'size_class']).size())

df_shai
```

```python
import cv2
import numpy as np
import os
import pandas as pd

# Function to draw bounding boxes using the explicit BX, BY, Width, Height values
def draw_boxes_from_coords(image_path, df_rows, output_path):
    # Read the image
    
    #jpg or png
    if image_path.endswith('.jpg'):
        img = cv2.imread(image_path)
    elif image_path.endswith('.png'):
        img = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)
    else:
        print(f"Unsupported image format: {image_path}")
        return False
    

    if img is None:
        print(f"Could not read image: {image_path}")
        return False
    
    # Draw boxes for all detections in this image
    for _, row in df_rows.iterrows():
        # Get bounding box coordinates
        bx = int(row['BX'])
        by = int(row['BY'])
        width = int(row['Width'])
        height = int(row['Height'])
        
        # Determine label (assuming we've classified as big or small)
        # If you haven't classified yet, use the approach from earlier
        
        label = row['size_class']
        
        
        # Set color based on label (BGR format)
        color = (0, 255, 0) if label == 'big' else (0, 0, 255)  # Green for big, Red for small
        
        # Draw rectangle
        cv2.rectangle(img, (bx, by), (bx + width, by + height), color, 2)
        
        # Add label
        cv2.putText(img, label, (bx, by - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)
    
    # Save the output image
    cv2.imwrite(output_path, img)
    return True

# Process each image
image_dir = '/Users/gilbenor/Library/CloudStorage/OneDrive-post.bgu.ac.il/thesisi/thesis document/Shai - exuviae'  # Update this path
output_dir = 'output_images'
os.makedirs(output_dir, exist_ok=True)






# Group by image_name to process all detections in each image
for image_name, group in df_shai.groupby('Label'):
    image_path = os.path.join(image_dir, image_name+'.jpg')
    output_path = os.path.join(output_dir, f"bbox_{image_name}.jpg")
    
    success = draw_boxes_from_coords(image_path, group, output_path)
    if success:
        print(f"Processed: {image_name}")
    else:
        print(f"Failed to process: {image_name}")
```

```python
import cv2
import numpy as np
import os
import pandas as pd
import json

# Function to draw both bounding boxes and keypoints on images
def draw_detections_on_image(image_path, bbox_rows, keypoint_file, output_path):
    # Read the image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Could not read image: {image_path}")
        return False
    
    # First, draw bounding boxes from df_shai
    if bbox_rows is not None:
        for _, row in bbox_rows.iterrows():
            # Get bounding box coordinates
            bx = int(row['BX'])
            by = int(row['BY'])
            width = int(row['Width'])
            height = int(row['Height'])
            
            # Determine label
            if 'size_class' in row:
                label = row['size_class']
            else:
                # Classify based on camera type and size
                if 'GX010191' in os.path.basename(image_path):
                    label = 'big' if width > 850 else 'small'
                else:
                    label = 'big' if width > 1300 else 'small'
            
            # Set color based on label (BGR format)
            color = (0, 255, 0) if label == 'big' else (0, 0, 255)  # Green for big, Red for small
            
            # Draw rectangle
            cv2.rectangle(img, (bx, by), (bx + width, by + height), color, 2)
            
            # Add label
            cv2.putText(img, label, (bx, by - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)
    
    # Next, draw keypoints if available
    if keypoint_file is not None and os.path.exists(keypoint_file):
        try:
            with open(keypoint_file, 'r') as f:
                keypoints_data = json.load(f)
            
            # Process keypoints data
            # Adjust this part based on the actual format of your keypoints files
            if isinstance(keypoints_data, list):
                for detection in keypoints_data:
                    keypoints = detection.get('keypoints', [])
                    # Draw each keypoint
                    for kp in keypoints:
                        x, y = int(kp[0]), int(kp[1])
                        confidence = kp[2] if len(kp) > 2 else 1.0
                        
                        if confidence > 0.01:  # Only draw high-confidence keypoints
                            cv2.circle(img, (x, y), 5, (255, 0, 0), -1)  # Blue circles for keypoints
            
            # If your keypoints format is different, adjust accordingly
        except Exception as e:
            print(f"Error processing keypoints file: {e}")
    
    # Save the output image
    cv2.imwrite(output_path, img)
    return True

# Main processing function
def process_images(df_shai, keypoints_dir, image_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    
    # Group by image_name to process all detections in each image
    for image_name, group in df_shai.groupby('Label'):
        image_path = os.path.join(image_dir, image_name+'.jpg')
        
        # Find corresponding keypoints file 
        # Adjust this based on how your files are named
        base_name = os.path.splitext(image_name)[0]
        keypoint_file = os.path.join(keypoints_dir, f"{base_name}.txt")  # Adjust extension if needed
        
        output_path = os.path.join(output_dir, f"combined_{image_name}.jpg")
        
        success = draw_detections_on_image(image_path, group, keypoint_file, output_path)
        if success:
            print(f"Processed: {image_name}")
        else:
            print(f"Failed to process: {image_name}")

# Usage
keypoints_dir = 'runs/pose/predict57/further_labels_files'  # Path to keypoints files
image_dir = '/Users/gilbenor/Library/CloudStorage/OneDrive-post.bgu.ac.il/thesisi/thesis document/Shai - exuviae'  # Path to original images
output_dir = 'output_combined'  # Path to save output images

# First, add size classification to df_shai if needed


# Process all images
process_images(df_shai, keypoints_dir, image_dir, output_dir)
```

```python
import cv2
import numpy as np
import os
import pandas as pd

# Function to parse keypoint files and draw on images
def draw_detections(image_path, keypoint_file, output_path, df_shai=None):
    # Read the image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Could not read image: {image_path}")
        return False
    
    image_height, image_width = img.shape[:2]
    
    # First try to draw from keypoint file
    if os.path.exists(keypoint_file):
        with open(keypoint_file, 'r') as f:
            lines = f.readlines()
        
        for line in lines:
            values = line.strip().split()
            if len(values) < 5:  # Need at least class_id and bbox coordinates
                continue
            
            # Parse the line
            class_id = int(values[0])
            
            # Parse normalized bbox coordinates and convert to pixel coordinates
            x_center = float(values[1]) * image_width
            y_center = float(values[2]) * image_height
            width = float(values[3]) * image_width
            height = float(values[4]) * image_height
            
            # Calculate bbox corners
            x1 = int(x_center - width/2)
            y1 = int(y_center - height/2)
            x2 = int(x_center + width/2)
            y2 = int(y_center + height/2)
            
            # Determine if big or small based on width (or from df_shai if available)
            # For this example, just using width threshold
            if 'GX010191' in os.path.basename(image_path):
                label = 'big' if width > 830 else 'small'
            else:
                label = 'big' if width > 1347 else 'small'
            
            # Set color based on label (BGR format)
            box_color = (0, 255, 0) if label == 'big' else (0, 0, 255)  # Green for big, Red for small
            
            # Draw bounding box
            cv2.rectangle(img, (x1, y1), (x2, y2), box_color, 2)
            cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, box_color, 2)
            
            # Parse and draw keypoints (each keypoint is x, y, conf)
            keypoints = []
            for i in range(5, len(values), 3):
                if i+2 < len(values):  # Make sure we have all three values
                    kp_x = float(values[i]) * image_width
                    kp_y = float(values[i+1]) * image_height
                    kp_conf = float(values[i+2])
                    
                    if kp_conf > 0.5:  # Only draw high-confidence keypoints
                        keypoints.append((int(kp_x), int(kp_y)))
                        cv2.circle(img, (int(kp_x), int(kp_y)), 5, (255, 0, 0), -1)  # Blue circles
            
            # Draw lines between keypoints if there are at least 2
            if len(keypoints) >= 2:
                for i in range(len(keypoints) - 1):
                    cv2.line(img, keypoints[i], keypoints[i+1], (255, 0, 255), 2)  # Magenta lines
    
    # Optionally, also draw bounding boxes from df_shai if provided
    if df_shai is not None:
        img_name = os.path.basename(image_path)
        img_name=img_name.replace('.jpg','')
        rows = df_shai[df_shai['Label'] == img_name]
        
        for _, row in rows.iterrows():
            if 'BX' in row and 'BY' in row and 'Width' in row and 'Height' in row:
                bx = int(row['BX'])
                by = int(row['BY'])
                width = int(row['Width'])
                height = int(row['Height'])
                
                # Determine label if available
                if 'size_class' in row:
                    label = row['size_class']
                else:
                    label = 'bx_by'

                # Draw with a different color to distinguish from keypoint boxes
                cv2.rectangle(img, (bx, by), (bx + width, by + height), (0, 165, 255), 2)  # Orange
                cv2.putText(img, f"{label} (shai)", (bx, by - 10), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 2)
                #draw both diagnonals
                cv2.line(img, (bx, by), (bx + width, by + height), (0, 165, 255), 2)
                #put txt numner of pixels  # Top-left to bottom-right
                cv2.putText(img, f"{row['Length']}", (bx, by - 10), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 2)
                cv2.line(img, (bx, by + height), (bx + width, by), (0, 165, 255), 2)
            

    # Save the output image
    cv2.imwrite(output_path, img)
    return True

# Process all images 
def process_all_images(keypoints_dir, image_dir, output_dir, df_shai=None):
    os.makedirs(output_dir, exist_ok=True)
    
    # Get all keypoint files
    keypoint_files = [f for f in os.listdir(keypoints_dir) if f.endswith('.txt')]
    
    for kp_file in keypoint_files:
        # Get base name without extension
        base_name = os.path.splitext(kp_file)[0]
        
        # Find corresponding image (might need to adjust extensions)
        image_file = f"{base_name}.jpg"
        if not os.path.exists(os.path.join(image_dir, image_file)):
            image_file = f"{base_name}.png"  # Try alternate extension
            if not os.path.exists(os.path.join(image_dir, image_file)):
                print(f"Could not find image for {kp_file}")
                continue
        
        image_path = os.path.join(image_dir, image_file)
        keypoint_path = os.path.join(keypoints_dir, kp_file)
        output_path = os.path.join(output_dir, f"annotated_{image_file}")
        
        success = draw_detections(image_path, keypoint_path, output_path, df_shai)
        if success:
            print(f"Processed: {image_file}")
        else:
            print(f"Failed to process: {image_file}")

    # Main execution
keypoints_dir = '/Users/gilbenor/Documents/code projects/msc/counting_research_algorithms/runs/pose/predict57/further_labels_files'
image_dir = '/Users/gilbenor/Library/CloudStorage/OneDrive-post.bgu.ac.il/thesisi/thesis document/Shai - exuviae'  # Path to original images  # Update to your images directory
output_dir = 'output_annotations'

# If you want to use df_shai for additional annotations:
# Add classification to df_shai first if needed
# df_shai['size_class'] = df_shai.apply(lambda row: 'big' if row['Width'] > 850 else 'small', axis=1)

# Process without df_shai
process_all_images(keypoints_dir, image_dir, output_dir,df_shai)

# Or process with df_shai if you want both annotations
# process_all_images(keypoints_dir, image_dir, output_dir, df_shai)
```
