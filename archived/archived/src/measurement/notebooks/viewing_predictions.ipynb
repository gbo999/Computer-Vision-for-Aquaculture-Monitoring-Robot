```python
#resize images for prediction
import os
import cv2
import numpy as np
import argparse
import tqdm

def resize_images(input_dir, output_dir, size):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    for root, dirs, files in os.walk(input_dir):
        for file in (files):
            if file.endswith(".png"):
                img = cv2.imread(os.path.join(root, file))
                img = cv2.resize(img, size)
                cv2.imwrite(os.path.join(output_dir, file), img)

resize_images("C:/Users/gbo10/OneDrive/measurement_paper_images/for imageJ/check/stabilized/car", "C:/Users/gbo10/OneDrive/measurement_paper_images/for imageJ/check/stabilized/car/carresized", (640, 640))
```
