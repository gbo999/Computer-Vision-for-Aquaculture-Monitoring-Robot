```python
from pathlib import Path

# Define the paths to your folders
folder1_path = Path(r'C:\Users\gbo10\OneDrive\measurement_paper_images\images used for imageJ\check\stabilized\shai\measurements/1')
folder2_path = Path(r'C:\Users\gbo10\Videos\data-science\Research-counting-algorithms\runs\pose\predict16')

def get_common_filename(filename):
    # Split the filename by '_' and return the first 4 parts joined by '_'
    parts = filename.stem.split('_')
    return '_'.join(parts[:2])

def list_files(folder):
    # List all files in the folder
    return [f for f in folder.iterdir() if f.is_file()]

def find_missing_images(folder1_files, folder2_files):
    # Extract common filenames
    folder1_common = set(get_common_filename(f) for f in folder1_files)
    folder2_common = set(get_common_filename(f) for f in folder2_files)
    
    # Find missing files
    missing_in_folder1 = folder2_common - folder1_common
    missing_in_folder2 = folder1_common - folder2_common
    
    return missing_in_folder1, missing_in_folder2

# List files in both folders
folder1_files = list_files(folder1_path)
folder2_files = list_files(folder2_path)

# Find missing images
missing_in_folder1, missing_in_folder2 = find_missing_images(folder1_files, folder2_files)

# Print the results
print("Missing in folder1:")
for filename in missing_in_folder1:
    print(filename)

print("\nMissing in folder2:")
for filename in missing_in_folder2:
    print(filename)

import csv  

# Save the results to csv file
csv_file_path = 'missing_images.csv'
with open(csv_file_path, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["Missing in folder1", "Missing in folder2"])
    
    # Write the missing filenames as columns
    max_len = max(len(missing_in_folder1), len(missing_in_folder2))
    missing_in_folder1 = list(missing_in_folder1)
    missing_in_folder2 = list(missing_in_folder2)
    
    for i in range(max_len):
        row = [
            missing_in_folder1[i] if i < len(missing_in_folder1) else '',
            missing_in_folder2[i] if i < len(missing_in_folder2) else ''
        ]
        writer.writerow(row)
```

```python
import os

def generate_filter_string(folder_path):
    # List all files in the folder
    filenames = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    
    # Extract relevant parts of the filename (e.g., 'GX010152_36_378') and remove the file extension
    filenames_no_ext = ['_'.join(f.split('.')[0].split('_')[1:4]) for f in filenames]
    
    # Generate the filter string
    filter_string = ' OR '.join([f'filename:{name}' for name in filenames_no_ext])
    
    return filter_string

# Define the path to your folder
folder_path = r'C:\Users\gbo10\OneDrive\measurement_paper_images\for fiftyone\resized\stretched\has_labels'

# Generate the filter string
filter_string = generate_filter_string(folder_path)

# Print the result
print(filter_string)
```
