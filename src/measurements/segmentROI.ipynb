```python
import cv2
import numpy as np

# Global variables
ref_point = []  # To store rectangle coordinates
crop = False

def click_and_crop(event, x, y, flags, param):
    # Grab references to the global variables
    global ref_point, crop

    # If the left mouse button was clicked, record the starting (x, y) coordinates
    if event == cv2.EVENT_LBUTTONDOWN:
        ref_point = [(x, y)]

    # Check to see if the left mouse button was released
    elif event == cv2.EVENT_LBUTTONUP:
        # Record the ending (x, y) coordinates and indicate that the cropping operation is finished
        ref_point.append((x, y))
        crop = True

        # Draw a rectangle around the region of interest
        cv2.rectangle(resized, ref_point[0], ref_point[1], (0, 255, 0), 2)
        cv2.imshow("image", resized)
        #zoom out
        

# Load the image, clone it, and setup the mouse callback function
img= cv2.imread('try.jpg')
screen_resolution = (1920, 1080)  # This is a common screen resolution. Adjust as needed.
scale_width = screen_resolution[0] / img.shape[1]
scale_height = screen_resolution[1] / img.shape[0]
scale = min(scale_width, scale_height)

# Resize the image
width = int(img.shape[1] * scale)
height = int(img.shape[0] * scale)
dim = (width, height)
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

clone = img.copy()
cv2.namedWindow("image")
cv2.setMouseCallback("image", click_and_crop)

# Keep looping until the 'q' key is pressed
while True:
    # Display the image and wait for a keypress
    cv2.imshow("image", resized)    
    #zooom out 

    key = cv2.waitKey(1) & 0xFF

    # If the 'r' key is pressed, reset the cropping region
    if key == ord("r"):
        image = clone.copy()

    # If the 'c' key is pressed, break from the loop
    elif key == ord("c"):
        break

# If there are two reference points, then crop the region of interest from the image and display it
if len(ref_point) == 2 and crop:
    roi = clone[ref_point[0][1]:ref_point[1][1], ref_point[0][0]:ref_point[1][0]]
    cv2.imshow("ROI", roi)
    cv2.waitKey(0)

    # Convert ROI to HSV
    hsv_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

    # Calculate the color range of the ROI
    lwr = hsv_roi.min(axis=(0, 1))
    upr = hsv_roi.max(axis=(0, 1))

    print(f"Lower HSV Bound: {lwr}")
    print(f"Upper HSV Bound: {upr}")

    # Now you can use lwr and upr for color segmentation in the original image or others.

# Close all open windows
cv2.destroyAllWindows()
```

```python
import cv2

def adjust_brightness_contrast(image, brightness=0, contrast=100):
    # The brightness value is added/subtracted to each pixel.
    # The contrast value should be in the range 0-100 where 100 is no change.
    # If you want to strengthen contrast, the value should be higher than 100.
    # OpenCV adds 1 by default to the contrast value for the calculation.
    
    # Scale contrast (contrast+1 is because the algorithm uses values from 1 to 131)
    contrast_scaled = contrast
    
    # Scale brightness
    # brightness_scaled = 127 * (brightness - 1) / 100
    
    # Apply the contrast and brightness to the image
    adjusted = cv2.convertScaleAbs(image, alpha=contrast_scaled)
    
    return adjusted

image=cv2.imread('try3.jpg')
adjusted=adjust_brightness_contrast(image, brightness=0, contrast=1.3)
#plot 
#resize window and keep resolution
cv2.imwrite('try3 copy.jpg',adjusted)




```
