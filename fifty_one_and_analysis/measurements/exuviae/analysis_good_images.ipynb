```python
import pandas as pd


df_length_analysis=pd.read_csv('/Users/gilbenor/Documents/code_projects/msc/counting_research_algorithms/fifty_one_and_analysis/measurements/exuviae/length_analysis_new_split.csv')

df_sample_tags=pd.read_csv('/Users/gilbenor/Documents/code_projects/msc/counting_research_algorithms/fifty_one_and_analysis/measurements/exuviae/spreadsheet_files/sample_tags.csv')




df_length_analysis['image_name'] = df_length_analysis['image_name'].str.split('colored_').str.get(-1)


df_sample_tags['image_name']=df_sample_tags['filepath'].str.split('.jpg').str.get(0).str.split('/').str.get(-1)





df_length_analysis_merged=df_length_analysis.merge(df_sample_tags, left_on='image_name', right_on='image_name')


df_length_analysis_merged
```

```python
df_length_new=df_length_analysis_merged.loc[df_length_analysis_merged['tags'].str.contains('good')]
df_length_new.loc[df_length_new['image_name'].str.contains('undistorted_GX010192_19_1095')]
```

```python
import pandas as pd

df_length_new['diff_big']=abs(df_length_new['total_length']-180)

df_length_new['diff_small']=abs(df_length_new['total_length']-145)


df_length_new['new_class'] = df_length_new.apply(lambda x: 'small' if  x['diff_small'] < x['diff_big'] and 'small mpe' in x['tags'] else 'big' if x['diff_big'] < x['diff_small'] and 'big mpe' in x['tags'] else None, axis=1)


nulls=df_length_new[['image_name','total_length', 'new_class','tags','labels']].loc[(df_length_new['new_class']!='big') & (df_length_new['new_class']!='small')].sort_values(by='total_length', ascending=False)

without_nulls=df_length_new.loc[(df_length_new['new_class']=='big') | (df_length_new['new_class']=='small')].sort_values(by='image_name', ascending=False)








```

```python
without_nulls.to_csv('/Users/gilbenor/Documents/code_projects/msc/counting_research_algorithms/fifty_one_and_analysis/measurements/exuviae/length_analysis_new_split_without_nulls.csv', index=False)
```
