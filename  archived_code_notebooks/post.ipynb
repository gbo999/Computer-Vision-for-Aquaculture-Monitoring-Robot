```python
import fiftyone as fo

dataset = fo.load_dataset("prawn_keypoints")
for sample in dataset:
    # Check detections
    if sample.detections is not None:
        for detection in sample.detections.detections:
            print("Detection fields:", detection.field_names)
            if hasattr(detection, "keypoints"):
                print("Detection keypoints type:", type(detection.keypoints))
                print("Detection keypoints value:", detection.keypoints)
                if len(detection.keypoints) > 0:
                    print("First keypoint entry type:", type(detection.keypoints[0]))
                    print("First keypoint entry value:", detection.keypoints[0])
                break
        break
    # Fallback: check sample.keypoints
    if hasattr(sample, "keypoints"):
        print("Sample keypoints type:", type(sample.keypoints))
        print("Sample keypoints value:", sample.keypoints)
        if len(sample.keypoints) > 0:
            print("First keypoint entry type:", type(sample.keypoints[0]))
            print("First keypoint entry value:", sample.keypoints[0])
        break
```
